```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Annotated, Literal
from langchain_core.messages import HumanMessage, BaseMessage
from langgraph.graph.message import add_messages
from langchain_groq import ChatGroq
from dotenv import load_dotenv, find_dotenv
```

```python
load_dotenv(find_dotenv())
```

```python
# model
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    streaming=True,
    temperature=0,
    max_retries=2
)
```

```python
# state
class Chat(TypedDict):
    messages:Annotated[list[BaseMessage],add_messages]
    repo_url: str
    repo_id: str
    intent: str
    retrieved_chunks: list
    repo_metadata: dict
    architecture_data: dict
    diagram_data: dict
    response: str
```

```python
# functions
def repo(state:Chat):
    return None

def tech_summary(state:Chat):
    return None

def non_tech_summary(state:Chat):
    return None

def architecture(state:Chat):
    return None

def system_design(state:Chat):
    return None

def chat_node(state:Chat):
    return None

def check_condition(state:Chat)->Literal['tech_summary','non_tech_summary','architecture','system_design','chat_node']:
    return state['intent']
```

```python
# graph creation
graph=StateGraph(Chat)
graph.add_node('repo',repo)
graph.add_node('tech_summary',tech_summary)
graph.add_node('non_tech_summary',non_tech_summary)
graph.add_node('architecture',architecture)
graph.add_node('system_design',system_design)
graph.add_node('chat_node',chat_node)

graph.add_edge(START,'repo')
graph.add_conditional_edges(
    "repo",
    check_condition,
    {
        "tech_summary": "tech_summary",
        "non_tech_summary": "non_tech_summary",
        "architecture": "architecture",
        "system_design": "system_design",
        "chat_node": "chat_node",
    },
)

graph.add_edge('tech_summary',END)
graph.add_edge('non_tech_summary',END)
graph.add_edge('architecture',END)
graph.add_edge('system_design',END)
graph.add_edge('chat_node',END)

workflow=graph.compile()
```
